```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

# Load datasets

```python
movies = pd.read_csv("/movies.csv")
ratings = pd.read_csv("/ratings.csv")
```

# Display datasets

```python
print(movies.head())
print(ratings.head())
```

# Create User-Movie Matrix

```python
movie_ratings = ratings.pivot_table(
    index='userId',
    columns='movieId',
    values='rating'
)
```

# Replace NaN with 0

```python
movie_ratings = movie_ratings.fillna(0)
```

# Calculate similarity between movies

```python
movie_similarity = cosine_similarity(movie_ratings.T)
```

# Create DataFrame

```python
movie_similarity_df = pd.DataFrame(
    movie_similarity,
    index=movie_ratings.columns,
    columns=movie_ratings.columns
)
```

# Recommendation Function

```python
def recommend_movies(movie_name):

    movie = movies[movies['title'] == movie_name]

    if movie.empty:
        print("Movie not found!")
        return

    movie_id = movie.iloc[0]['movieId']

    similar_scores = movie_similarity_df[movie_id]

    similar_movies = similar_scores.sort_values(
        ascending=False
    )[1:6]

    print(f"\nRecommended movies for '{movie_name}':\n")

    for mid in similar_movies.index:
        title = movies[movies['movieId'] == mid]['title'].values
        if len(title) > 0:
            print(title[0])
```

# Example

```python
recommend_movies("Toy Story (1995)")
```
